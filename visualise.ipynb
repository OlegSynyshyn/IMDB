```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IMDB-Movie-Data.csv")

df.info()
```

```python
df['Rating'].plot(kind= 'hist', bins=15)
plt.show()
```

```python
df.plot(kind='scatter', x='Rating', y='Votes')
```

```python
year = df['Year'].value_counts().nlargest(20).plot(kind='pie', autopct='%1.f%%')
year.set_title("    2006 - 2016")
```

```python
# Runtime (Minutes) Year
mean1 = df.groupby('Year')['Runtime (Minutes)'].mean()
p = mean1.plot(kind='barh', grid=True)
p.set_title("    ")
p.set_xlabel("  ")
p.set_ylabel("")

```

```python
# Revenue (Millions), Rating
df.plot(kind='scatter', x='Rating', y='Revenue (Millions)')
```
